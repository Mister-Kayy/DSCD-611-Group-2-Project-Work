```python
# PHASE 3: AVERAGE RAINFALL FOR ALL LOCATIONS
print("="*60)
print("CALCULATING AVERAGE RAINFALL FOR ALL LOCATIONS")
print("="*60)

# We'll calculate 3 things for each location:
# 1. Average monthly rain
# 2. Maximum monthly rain (biggest flood potential)
# 3. Number of flood months

print("\n" + "-"*70)
print(f"{'LOCATION':25} | {'AVG RAIN':>8} | {'MAX RAIN':>8} | {'FLOOD MONTHS':>12}")
print("-"*70)

# Dictionary to store results
rain_stats = {}

for location in sorted(df_recent['Name'].unique()):
    # Get data for this location
    loc_data = df_recent[df_recent['Name'] == location]

    # Calculate statistics
    avg_rain = loc_data['monthly_total_rfh'].mean()
    max_rain = loc_data['monthly_total_rfh'].max()
    flood_months = loc_data['exceeds_2yr'].sum()  # Count 1's in exceeds_2yr column

    # Store for later use
    rain_stats[location] = {
        'avg_rain': avg_rain,
        'max_rain': max_rain,
        'flood_months': flood_months
    }

    # Print in a nice table
    print(f"{location:25} | {avg_rain:8.1f} | {max_rain:8.1f} | {flood_months:12d}")

print("-"*70)

# Find the wettest and driest locations
wettest = max(rain_stats.items(), key=lambda x: x[1]['avg_rain'])
driest = min(rain_stats.items(), key=lambda x: x[1]['avg_rain'])
most_floods = max(rain_stats.items(), key=lambda x: x[1]['flood_months'])

print(f"\n📊 QUICK ANALYSIS:")
print(f"  • Wettest location: {wettest[0]} ({wettest[1]['avg_rain']:.1f}mm avg)")
print(f"  • Driest location: {driest[0]} ({driest[1]['avg_rain']:.1f}mm avg)")
print(f"  • Most flood months: {most_floods[0]} ({most_floods[1]['flood_months']} months)")
```

```python
# PHASE 4: CREATE FLOOD RISK SCORE
print("="*60)
print("CREATING FLOOD RISK SCORE FOR EACH LOCATION")
print("="*60)

print("\n📝 HOW WE CALCULATE FLOOD RISK:")
print("1. Average Rainfall (30% weight) - more rain = more risk")
print("2. Maximum Rainfall (40% weight) - big floods matter most")
print("3. Flood Frequency (30% weight) - how often floods happen")
print("   → Combined into a 0-100 risk score")

print("\n" + "-"*85)
print(f"{'LOCATION':25} | {'AVG RAIN':>8} | {'MAX RAIN':>8} | {'FLOODS':>6} | {'RISK SCORE':>10} | {'RISK LEVEL':>12}")
print("-"*85)

# We'll normalize each factor to 0-100 scale
def normalize_scores(values):
    """Convert values to 0-100 scale"""
    min_val = min(values)
    max_val = max(values)
    if max_val == min_val:  # Avoid division by zero
        return [50 for _ in values]  # All middle if all same
    return [(v - min_val) / (max_val - min_val) * 100 for v in values]

# Extract data for normalization
locations = []
avg_rains = []
max_rains = []
flood_counts = []

for location in sorted(df_recent['Name'].unique()):
    loc_data = df_recent[df_recent['Name'] == location]

    locations.append(location)
    avg_rains.append(loc_data['monthly_total_rfh'].mean())
    max_rains.append(loc_data['monthly_total_rfh'].max())
    flood_counts.append(loc_data['exceeds_2yr'].sum())

# Normalize each factor
avg_scores = normalize_scores(avg_rains)      # 0-100
max_scores = normalize_scores(max_rains)      # 0-100
flood_scores = normalize_scores(flood_counts) # 0-100

# Calculate weighted risk score (30% avg, 40% max, 30% floods)
risk_scores = []
for i in range(len(locations)):
    risk = (avg_scores[i] * 0.3) + (max_scores[i] * 0.4) + (flood_scores[i] * 0.3)
    risk_scores.append(risk)

# Display results
for i, location in enumerate(locations):
    risk = risk_scores[i]

    # Determine risk level
    if risk >= 80:
        risk_level = "VERY HIGH"
    elif risk >= 60:
        risk_level = "HIGH"
    elif risk >= 40:
        risk_level = "MEDIUM"
    elif risk >= 20:
        risk_level = "LOW"
    else:
        risk_level = "VERY LOW"

    print(f"{location:25} | {avg_rains[i]:8.1f} | {max_rains[i]:8.1f} | {flood_counts[i]:6d} | {risk:10.1f} | {risk_level:>12}")

print("-"*85)

# Find top 3 riskiest locations
risk_data = list(zip(locations, risk_scores))
risk_data.sort(key=lambda x: x[1], reverse=True)

print(f"\n🏆 TOP 3 RISKIEST LOCATIONS:")
for rank, (location, score) in enumerate(risk_data[:3], 1):
    print(f"  {rank}. {location} (Score: {score:.1f})")

print(f"\n🛡️  TOP 3 SAFEST LOCATIONS:")
for rank, (location, score) in enumerate(risk_data[-3:], 1):
    print(f"  {rank}. {location} (Score: {score:.1f})")
```

```python
# CREATE MAPPING DATASET FROM SCRATCH
print("="*60)
print("CREATING COMPLETE MAPPING DATASET")
print("="*60)

# 1. Get all unique locations
locations = sorted(df_recent['Name'].unique())

# 2. Calculate statistics for each location
location_data = []

for location in locations:
    loc_data = df_recent[df_recent['Name'] == location]

    # Calculate statistics
    avg_rain = loc_data['monthly_total_rfh'].mean()
    max_rain = loc_data['monthly_total_rfh'].max()
    flood_months = loc_data['exceeds_2yr'].sum()

    # Get PCODE
    pcode = loc_data['PCODE'].iloc[0]

    location_data.append({
        'PCODE': pcode,
        'Name': location,
        'Avg_Rain_mm': avg_rain,
        'Max_Rain_mm': max_rain,
        'Flood_Months': flood_months
    })

# Convert to DataFrame
map_data = pd.DataFrame(location_data)

# 3. Add exact coordinates
exact_coords = {
    'Ada East': (5.7834, 0.6329),
    'Ada West': (5.8672, 0.2938),
    'Adenta Municipal': (5.7231, -0.1651),
    'Ayawaso West': (5.6203, -0.1904),
    'Ga Central Municipal': (5.6267, -0.2792),
    'Ga East': (5.7346, -0.2030),
    'Ga South Municipal': (5.5705, -0.3400),
    'Ga West Municipal': (5.7039, -0.3013),
    'Kpone Katamanso': (5.6914, 0.0624),
    'La Dade-Kotopon': (5.5600, -0.1500),
    'La-Nkwantanang-Madina': (5.6719, -0.1653),
    'Ledzokuku Municipal': (5.5841, -0.1062),
    'Ningo/Prampram': (5.7139, 0.1078),
    'Shai Osudoku': (5.8824, -0.1116),
    'Tema West Municipal': (5.6372, -0.0194)
}

# Add coordinates to map_data
map_data['Latitude'] = map_data['Name'].map(lambda x: exact_coords[x][0])
map_data['Longitude'] = map_data['Name'].map(lambda x: exact_coords[x][1])

# 4. Calculate risk scores
# Normalize function
def normalize_scores(values):
    min_val = min(values)
    max_val = max(values)
    if max_val == min_val:
        return [50 for _ in values]
    return [(v - min_val) / (max_val - min_val) * 100 for v in values]

# Get values for normalization
avg_rains = map_data['Avg_Rain_mm'].tolist()
max_rains = map_data['Max_Rain_mm'].tolist()
flood_counts = map_data['Flood_Months'].tolist()

# Normalize
avg_scores = normalize_scores(avg_rains)
max_scores = normalize_scores(max_rains)
flood_scores = normalize_scores(flood_counts)

# Calculate weighted risk score (30% avg, 40% max, 30% floods)
risk_scores = []
for i in range(len(map_data)):
    risk = (avg_scores[i] * 0.3) + (max_scores[i] * 0.4) + (flood_scores[i] * 0.3)
    risk_scores.append(risk)

map_data['Risk_Score'] = risk_scores

# Add risk level
def get_risk_level(score):
    if score >= 80:
        return "VERY HIGH"
    elif score >= 60:
        return "HIGH"
    elif score >= 40:
        return "MEDIUM"
    elif score >= 20:
        return "LOW"
    else:
        return "VERY LOW"

map_data['Risk_Level'] = map_data['Risk_Score'].apply(get_risk_level)

print("✅ MAPPING DATASET CREATED FROM SCRATCH!")
print(f"Contains: {len(map_data)} locations")
print("\n📋 COMPLETE DATASET:")
print(map_data.to_string(index=False))
```

```python
import folium
import geopandas as gpd
import requests
import os
from folium.features import DivIcon

# --- 1. AUTOMATIC DOWNLOADER ---
def download_adm2_shapefile():
    url = "https://github.com/wmgeolab/geoBoundaries/raw/main/releaseData/gbOpen/GHA/ADM2/geoBoundaries-GHA-ADM2.geojson"
    filename = "geoBoundaries-GHA-ADM2.geojson"

    if not os.path.exists(filename):
        print(f"⬇️ Downloading ADM2 Shapefile from {url}...")
        try:
            r = requests.get(url)
            r.raise_for_status()
            with open(filename, 'wb') as f:
                f.write(r.content)
            print("✅ Download Successful!")
        except Exception as e:
            print(f"❌ Download failed: {e}")
            return None
    else:
        print("✅ File already exists in notebook.")

    return filename

# --- 2. GET DATA & SETUP ---
shapefile_path = download_adm2_shapefile()

image_palette = {
    'VERY HIGH': '#FF0000', 'HIGH': '#0000FF',
    'MEDIUM': '#32CD32', 'LOW': '#FFD700', 'VERY LOW': '#D3D3D3'
}

if shapefile_path:
    # Load the shapefile
    gdf = gpd.read_file(shapefile_path)

    # --- 3. MATCHING LOGIC ---
    def find_match(geo_name, user_df):
        # Clean names for better matching
        clean_geo = str(geo_name).lower().replace(' district', '').replace(' municipal', '').replace(' metropolitan', '').strip()

        for idx, row in user_df.iterrows():
            clean_user = str(row['Name']).lower().replace(' district', '').replace(' municipal', '').replace(' metropolitan', '').strip()
            # Match check
            if clean_geo == clean_user or clean_user in clean_geo:
                return row
        return None

    # Prepare data for map
    gdf['Risk_Level'] = 'VERY LOW'
    gdf['Risk_Score'] = 0
    gdf['User_Name'] = ''

    # Run Match
    name_col = 'shapeName' if 'shapeName' in gdf.columns else gdf.columns[0]
    matched_count = 0

    for idx, row in gdf.iterrows():
        match = find_match(row[name_col], map_data)
        if match is not None:
            gdf.at[idx, 'Risk_Level'] = match['Risk_Level']
            gdf.at[idx, 'Risk_Score'] = match['Risk_Score']
            gdf.at[idx, 'User_Name'] = match['Name']
            matched_count += 1

    print(f"✅ Matched {matched_count} districts from your data.")

    # --- 4. CREATE PRECISE MAP ---
    # Filter to only matched districts (Greater Accra)
    plot_gdf = gdf[gdf['User_Name'] != ''].copy()

    if len(plot_gdf) > 0:
        precise_map = folium.Map(location=[5.65, -0.15], zoom_start=10, tiles='CartoDB positron')

        # Add Polygons
        folium.GeoJson(
            plot_gdf,
            style_function=lambda feature: {
                'fillColor': image_palette.get(feature['properties']['Risk_Level'], 'gray'),
                'color': 'black', 'weight': 1, 'fillOpacity': 0.7,
            },
            tooltip=folium.GeoJsonTooltip(
                fields=['User_Name', 'Risk_Level'], aliases=['District:', 'Risk:']
            )
        ).add_to(precise_map)

        # Add Markers
        for idx, row in plot_gdf.iterrows():
            center = row.geometry.centroid
            html_icon = """
            <div style="position: relative; width: 18px; height: 18px;
                background-color: #FF69B4; border-radius: 50%; border: 1px solid black;
                display: flex; align-items: center; justify-content: center;
                font-weight: bold; font-size: 14px; color: black;">+</div>
            """
            folium.Marker(
                location=[center.y, center.x],
                icon=DivIcon(html=html_icon, icon_size=(18, 18), icon_anchor=(9, 9))
            ).add_to(precise_map)

        # Add Legend
        legend_html = '''
        <div style="position: fixed; bottom: 30px; right: 30px; width: 180px;
             background-color: #dcb372; border: 2px solid black; z-index:9999; padding: 10px;
             font-family: Arial, sans-serif; font-size:12px; box-shadow: 3px 3px 5px rgba(0,0,0,0.3);">
             <h4 style="margin:0 0 8px 0; border-bottom: 1px solid black;">Legend</h4>
             <div style="margin-bottom: 5px;">
                <span style="display:inline-block; width:15px; height:15px; background:#FF69B4;
                      border:1px solid black; border-radius:50%; text-align:center; line-height:15px; font-weight:bold;">+</span>
                Flooded communities
             </div>
             <b>Flood hazard zone</b><br>
             <div style="display: flex; align-items: center; margin-top: 4px;">
                <div style="background: #FF0000; width: 15px; height: 15px; margin-right: 5px; border:1px solid black;"></div> Very high risk
             </div>
             <div style="display: flex; align-items: center; margin-top: 4px;">
                <div style="background: #0000FF; width: 15px; height: 15px; margin-right: 5px; border:1px solid black;"></div> High risk
             </div>
             <div style="display: flex; align-items: center; margin-top: 4px;">
                <div style="background: #32CD32; width: 15px; height: 15px; margin-right: 5px; border:1px solid black;"></div> Medium risk
             </div>
             <div style="display: flex; align-items: center; margin-top: 4px;">
                <div style="background: #FFD700; width: 15px; height: 15px; margin-right: 5px; border:1px solid black;"></div> Low risk
             </div>
        </div>
        '''
        precise_map.get_root().html.add_child(folium.Element(legend_html))

        display(precise_map)
        precise_map.save('accra_district_boundaries_clean.html')
    else:
        print("❌ Loaded shapefile but found no matches. Check district names.")
else:
    print("❌ Could not download file. Please check internet connection.")
```
